### 3. Visualización de Datos
Usando el diagrama de dispersión o dispersograma de la libreria seaborn (sns), mostraremos como se relacionan dos variables entre sí.  En este caso analizaremos como se relacionan las variables Age (edad) con los Expenditures (Gastos). 

```python
sns.scatterplot(data=test_df, x="Age", y="Expenditures")
```

En la grafica encontramos que a medida que tenemos una mayor edad, tendemos a un mayor nivel de gastos.

```python
sns.scatterplot(data=test_df, x="Age", y="Expenditures", hue="Age Cohort", alpha=0.3)  
plt.legend(bbox_to_anchor=(1.02, 1),loc='upper left', borderaxespad=0)
```

En esta grafica los individuos son representados en diferentes grupos de edades mediante diferentes colores, usando la variable "Age Cohort" en el parametro hue. Alpha por su parte es usado para establecer la opacidad de los puntos. 

```python
Adolescentes = test_df[test_df["Age Cohort"] == "13 to 17"]
Adultos = test_df[test_df["Age Cohort"] == "22 to 50"]
Infantes = test_df[test_df["Age Cohort"] == "0 to 5"]
Preadultos = test_df[test_df["Age Cohort"] == "18 to 21"]
Cincuentones = test_df[test_df["Age Cohort"] == "51+"]
Niños = test_df[test_df["Age Cohort"] == "6 to 12"]
```

Del conjunto de los 1.000 datos en test_df, usando la columna "Age Cohort" realizaremos diferentes filtros para seleccionar grupos de datos segun las edades.

```python
sns.lmplot(x="Age", y="Expenditures", hue="Age Cohort", data=test_df)
```

la funcion lmplot, adicional a crear el diagrama de dispersión entre la dos variables, nos permite trazar una linea de regresión para cada una de los grupos o "Age Cohourt" en nuestor caso.




En las graficas que sigue se profundiza en el analisis de los gastos para cada una de los grupos de edades establecidos previamente, pero ahora detalla por grupo etnico, iniciando con los infantes.

```python
sns.lmplot(x="Age", y="Expenditures", hue="Ethnicity", data=Infantes)
```
