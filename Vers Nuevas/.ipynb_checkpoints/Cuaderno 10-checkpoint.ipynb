Debes asegurarte de llamar las librerias necesarias para el desarrollo de tus actividades, de tal forma que no te generen errores o inconsistencias en pasos posteriores.
En este caso adicionamos parametros para mejorar la resolución grafíca

```python
from scipy.stats import pearsonr      # Libreria de funciones estadisticas                   
import pandas as pd                   # Librería para gestionar archivos
import matplotlib.pyplot as plt       # Librería para resumir datos con gráficos
import seaborn as sns                 # Librería para resumir datos con gráficos
# import seaborn as sns; sns.set(style="ticks", color_codes=True)
plt.rcParams['figure.dpi'] = 150      # Para mejorar la resolución gráfica
plt.rcParams['savefig.dpi'] = 150     # Para mejorar la resolución gráfica
```

### 2. Conjunto de datos a utilizar

Utilizamos los datos usados en el cuaderno anterior (newdata.csv).
Recuerda que se trata de un archivo csv o de datos separados por comas.
Con la opción datos.head() vamos a observar los primeros cinco registros o filas del archivo de datos. 

```python
datos = pd.read_csv("newdata.csv")
datos.head()
```

### 3. Analisis de Variables

Ahora se creara un subconjunto data, a partir del conjunto de datos, de las variables que se queren analizar, que en este caso son: Cost Delivery, Number of Comments, Expected Delivery Time y Time(min).

```python
data = datos[['Cost Delivery', 'Number of Comments', 'Expected Delivery Time', 'Time(min)']]
data.head()
```

Sin embargo, debemos asegurarnos que los todos los datos sean valores validos, es decir que no contengan valores nulos, lo cual constatamos con el siguiente comando:

```python
data.isnull().sum()
```

Mediante el siguiente comando eliminanos todas las instancias o filas de datos que contienen datos nulos o invalidos:

```python
data = data.dropna()
```

Validamos que no tengamos ningun dato nulo para proceder a buscar la relacion entre las variables de nuestro analisis.

```python
data.isnull().sum()
```

Ahora, vamos a ver cómo en Python puede desarrollarse una función que nos calcule la correlación entre dos variables para cuantificar con un número la covariación entre dos variables y luego usar ese número dentro de un gráfico que nos permita tener un panorama de la covariación bivariada entre tres o más pares de variables.

```python
def corrfunc(x, y, ax=None, **kws):
    r, _ = pearsonr(x, y)
    ax = ax or plt.gca()
    ax.annotate(f'r = {r:.2f}', xy=(.1, .9), xycoords=ax.transAxes) 
```

Mediante esta funcion encontraremos la relacion que hay entre cada par de variables, conocido como el coeficiente de correlación de Pearson.

Este es una medida de dependencia lineal entre dos variables aleatorias cuantitativas. Nos permite medir el grado de relación de dos variables siempre y cuando ambas sean cuantitativas y continuas.

La siguiente secuencia de comandos muestra la relación bivariada de las variables de negocio

(ten presente que la complejidad de los calculos, puede tardar algunos minutos dependiendo de la capacidad de tu maquina).

```python
g = sns.PairGrid(data)
g.map_upper(plt.scatter, s=10)
g.map_diag(sns.histplot, kde=True)
g.map_lower(corrfunc)
g.map_lower(sns.kdeplot, cmap="Blues_d")
```

Una recomendación para el gráfico es acortar los nombres de las variables, entonces lo haremos de inmediato así

```python
datas = data.rename(columns={'Cost Delivery': 'CD', 'Number of Comments': 'NC', 'Expected Delivery Time': 'EDT', 'Time(min)': 'CT'})
datas.head()
```

```python
datas = datas.dropna()
datas.head()
```

```python
g = sns.PairGrid(datas, diag_sharey=False, corner=True)
g.map_diag(sns.histplot, kde=True)
g.map_lower(corrfunc)
g.map_lower(sns.kdeplot, cmap="Blues_d")
```

Aunque volvimos a hacer el gráfico, ahora observamos que la variable NC (number of comments) no aparece graficada. Para entender la razón podemos ver qué tipo de variable es

```python
datas.dtypes
```

La variable NC es de tipo object. Debemos configurarla como variable de tipo float64 para que al hacer el gráfico pueda aparecer.

```python
datas['NC'] = datas['NC'].astype(float, errors = 'raise')
datas.dtypes
```

```python
g = sns.PairGrid(datas, diag_sharey=False, corner=True)
g.map_diag(sns.histplot, kde=True)
g.map_lower(corrfunc)
g.map_lower(sns.kdeplot, cmap="Blues_d")
```
