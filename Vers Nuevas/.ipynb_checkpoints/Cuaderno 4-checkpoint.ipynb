# Cuaderno 4: El Cuarteto de Anscombe

El cuarteto de Anscombe (cuyo nombre se debe a Francis Anscombe) se refiere a cuatro tablas de datos. Estas cuatro tablas comparten literalmente propiedades estadísticas idénticas si se les mira con un resumen estadístico en forma de tabla, pero manifiestan comportamientos estadísticos totalmente diferentes si se les mira con un gráfico.

Cada tabla de datos tiene dos variables (x, y) y cada variable tiene once observaciones. El cuarteto es una demostración de la importancia de mirar gráficamente un conjunto de datos antes de analizarlos. Recuerda que para un buen analisis son fundamental los conceptos de TABULACIÓN y GRAFICACIÓN, uno solo de ellos no es su ficiente para la comprensión de los datos

Vamos a mirar los datos y luego vamos a resumirlos a través de tablas y de gráficos para aprender nuestra lección descrita en el capítulo 4 del libro.

### 1. Llamado a las librerias
LLamado a las librerias necesarias para el desarrollo de tus actividades.
En este caso adicionamos parametros para mejorar la reoslucion grafica

```python
import pandas as pd                   # Libreria especializada en el manejo y estructura de datos 
import seaborn as sns                 # Librería para resumir datos con gráficos
import matplotlib as plt              # Librería para gráficos
plt.rcParams['figure.dpi'] = 300      # Para mejorar la resolución gráfica
plt.rcParams['savefig.dpi'] = 300     # Para mejorar la resolución gráfica
```

### 2. Conjunto de datos o datasets a utilizar
En este cso utilizaremos el archivo "CuartetoAnscombe.csv". 
Recuerda que se trata de un archivo csv pero en este caso tienen una particularidad son separados por tabulador (no por coma como habia ocurrido previamente), razón por la cual aparece el parametro sep (que significa separador) y '\t' indica que estan seprados por el tabulador.
Como de costumbre asegurate de validar que los datos han sido leidos en forma correcta

```python
Anscombe = pd.read_csv("CuartetoAnscombe.csv", sep ='\t') # En Anscombe se guardara la informacion del archivo de datos "CuartetoAnscombe.csv"
Anscombe.head()                                          # Muestra el contenido de los cinco primeros registros del dataframe "Anscombe"
```

```python
Anscombe.describe()              # Muestra las principales estadisticas descriptivas de los datos "Anscombe"
```

Observa que en la tabla resultante de la sintaxis anterior, los siguientes valores son idénticos para las parejas (x,y): 
* count (número de observaciones)
* mean (promedio)
* std (desviación estándar)

### 3. Analisis de Datos.
Vamos a profundizar en el analisis de cada par de variables X, Y para ver como se relacionan entre ellas. Es decir vamos a indagar como todos los valores de X1 y todos los valores de Y1 se relacionan entre si, en un procedimiento que en la estadistica se conoce como un coeficiente de correlacion, que lo que cuantifica es la relacion lineal entre esas dos variables. Simplificandolo es ver si la relacion entre esas dos variables se aproxima a una linea recta imaginaria trazada entre los datos.
Lo realizaremos para cada par de variables X1-Y1, X2-Y2, X3-Y3 y X4-Y4.

```python
print('Correlación Pearson X1-Y1: ', Anscombe['x1'].corr(Anscombe['y1'], method='pearson'))
print('Correlación Pearson X2-Y2: ', Anscombe['x2'].corr(Anscombe['y2'], method='pearson'))
print('Correlación Pearson X3-Y3: ', Anscombe['x3'].corr(Anscombe['y3'], method='pearson'))
print('Correlación Pearson X4-Y4: ', Anscombe['x4'].corr(Anscombe['y4'], method='pearson'))
```

Observa los resultados!!!!
Este coeficiente mas conocido como coeficiente de correlación Pearson es matemáticamente igual (con algunas variaciones en las últimas décimas) para los cuatro conjuntos. Sin embargo observa lo que ocurre cuando obtenemos el diagrama de dispersion o dispersigrama (scatterplot) de cada conjunto de datos. Pon especial atención en cómo se distribuyen los punticos azules dentro de cada plano cartesiano. ¿Que observas?

```python
sns.lmplot(x="x1", y="y1", data=Anscombe)        # Crea el diagrama de dispersion y la linea de regresion entre x1-y1
```

```python
sns.lmplot(x="x2", y="y2", data=Anscombe)      # Crea el diagrama de dispersion y la linea de regresion entre x2-y2
```

```python
sns.lmplot(x="x3", y="y3", data=Anscombe)             # Crea el diagrama de dispersion y la linea de regresion entre x3-y3
```

```python
sns.lmplot(x="x4", y="y4", data=Anscombe)         # Crea el diagrama de dispersion y la linea de regresion entre x4-y4
```

En esta parte cargaremos un archivo con los mismos datos X-Y, pero ahora solo le indicaremos el conjunto al que pertenece cada  par de variables, asi por ejemplo x1-y1 se indicara que pertenece al conjunto 1, x2-y2 al conuto 2 y asi sucesivamnete. Esto con el proposito de facilitar la comparación visual de los resultados. Al igual que el anterior estan separados por tabulador.

```python
Anscombe_B = pd.read_csv("Anscombe(FormaB).csv", sep ='\t')   # En Anscombe_B se guardara la informacion del archivo de datos "Anscombe(FormaB).csv"
```

Para la visualizacion utilizaremos el comnando FacetGrid que permite la creación de multiples graficas en un solo eje, proceso que se conoce como subtramas. En este caso caso le estamos indicando que las tramas o graficas las realice basados en la variable Conjunto, la cual corresponde a los grupos x1-y1, x2-y2, x3-y3 y x4-y4.
Despues con el comando map le indicamos que grafique los diagramas de dispersion de los grupos correspondientes a la variable Conjunto, usando como ejes x,y.
El parametro alpha nos permite establecer el nivel de opcaidad de los puntos en la grafica y legend que aprezca el titulo correspondiente al conjunto que cada grafico representa.


```python
g = sns.FacetGrid(Anscombe_B, col="Conjunto")   # Definicion de la variable Conjunto como parametro de las subtramas de los datos x,y
g.map(sns.scatterplot, "x", "y", alpha=.7)      # Diagramas de dispersión de cada conjunto.
g.add_legend()                                  # Adición del titulo de cada diagrama de dispersión
```
