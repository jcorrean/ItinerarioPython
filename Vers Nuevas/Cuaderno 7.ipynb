### 2. Llamado a las librerias
Como es costumbre te recordamos la importancia de llamar las librerias que utilizaras para el desarrollo de tu actividad.

```python
import wikipedia                                           # llamado a la libreria wikipedia
import numpy as np                                         # libreria 
import matplotlib.pyplot as plt                            # Libreria para graficación
from wordcloud import WordCloud, STOPWORDS
```

Con la siguiente sintaxis, vamos a extraer el texto relacionado con web scraping disponible en wikipedia.

### 3. Extraer la Informacion 
En la siguiente secuencia de instrucciones extraemos la informacion de wikipedia

```python
wiki = wikipedia.page('Data Analytics')                    # Se establece el termino que se buscara en Wikipedia
```

```python
texto = wiki.content                                       # En texto se trae la información de wikipedia sobre'Data Analytics'
```

```python
texto                                                      # Muestra el texto extraido de Wikipedia
```

### 4. Generando la bube de palabras
La dimensión de la nube de palabras se puede cambiar, modificando los parametros width y height. Mediante el parametro random_state se garantiza la reproducibilidad de la nube, es decir que siempre se genere la misma.

El background_colour es el color que se quiere para el fondo de la nube, mientras que el colormap brinda las opciones para las combinaciones de los textos. Algunas opciones son 'Dark2','Greys','Oranges','Pastel1', 'bone', 'inferno','plasma','summer', 'turbo','viridis' y 'winter'. Si desea conocer a fondo mas opciones se reocmienda al lector revisar la documentacion de la libreria Matplotlib en https://matplotlib.org/

El parametro Collocations asegura la no duplicidad de palabras con la opcion "False". Finalmente el parametro stopword hace referencia a las palabras comunes que proporcionan poco valor al significado del texto. Usamos STOPWORDS de la libreria wordcloud como se indicó en la parte 2 de la actividad.

```python
Nube = WordCloud(width= 2500, height = 2000, random_state=1, background_color='red', colormap='plasma', collocations=False, stopwords = STOPWORDS).generate(texto)
plt.figure(figsize=(15, 15))                                      # Tamaño de la figura
plt.imshow(Nube)                                                  # Muestra la imagen
plt.axis("off")                                                   # Sin detalles del eje
```

En el ejemplo que sigue, se cambiara el tamaño y los atributo de color de la nube de palabras

```python
Nube1 = WordCloud(width= 1500, height = 1000, random_state=1, background_color='blue', colormap='winter', collocations=False, stopwords = STOPWORDS).generate(texto)
plt.figure(figsize=(12, 10))                                     # Tamaño de la figura
plt.imshow(Nube1)                                                # Muestra la imagen
plt.axis("off")                                                  # Sin detalles del eje
```

La nube de palabras tambien ppuede tener formas especificas usando las capacidades de la libreria numpy.

```python
x, y = np.ogrid[:300, :300]                                 #  Eje de referencia
mascara = (x - 150) ** 2 + (y - 150) ** 2 > 130 ** 2        # Calculo de la mascara para mostrar el circulo
mascara = 255 * mascara.astype(int)
Nube2 = WordCloud(background_color="white", repeat=True, mask=mascara)  # generación de la nube
Nube2.generate(texto)

plt.axis("off")                                             # Sin detalles del eje
plt.imshow(Nube2, interpolation="bilinear")                 # Muestra la imagen
plt.show()                                                  
```
